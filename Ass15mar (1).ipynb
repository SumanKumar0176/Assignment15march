```python



Q3- What is unsupervised learning? List some examples of unsupervised learning.
Ans-
Unsupervised learning is a type of machine learning where a model is trained on an unlabeled dataset,
 without any corresponding output labels. The goal of unsupervised learning is to learn patterns, relationships,
 and structure within the data, without any explicit supervision.

Some examples of unsupervised learning are:
1.Clustering - Given an unlabeled dataset, the task is to group similar data points together into clusters, based on
 some similarity metric. This can be useful for market segmentation, image segmentation, or anomaly detection.

2.Dimensionality reduction - Given a high-dimensional dataset, the task is to find a low-dimensional representation 
of the data that preserves most of the relevant information. This can be useful for visualization, feature extraction,
 or data compression.

3.Association rule mining - Given a dataset of transactions, the task is to discover rules that describe frequent 
co-occurrences of items in the data, such as "customers who buy diapers are also likely to buy baby wipes". This can 
be useful for recommendation systems, market basket analysis, or cross-selling.

4.Anomaly detection - Given a dataset of normal behavior, the task is to detect rare and unusual instances that 
deviate from the norm. This can be useful for fraud detection, intrusion detection, or quality control.

5.Generative modeling - Given a dataset of samples, the task is to learn a model that can generate new samples
 that resemble the original data distribution. This can be useful for image synthesis, text generation, or data augmentation.
```

```python
Q7- How can unsupervised learning be used in anomaly detection?
Ans-
Unsupervised learning can be used for anomaly detection in the following way:

1.Train an unsupervised learning model (such as clustering algorithms or dimensionality reduction techniques)
 on a dataset that contains both normal and anomalous instances.

2.The unsupervised learning model learns the patterns in the normal instances and forms a representation of 
what is considered normal in the data.

3.When the model is presented with a new, unseen instance, it compares the features of the instance to the
 learned normal patterns. If the instance is significantly different from the learned patterns, the model 
 identifies it as an anomaly.

4.The identified anomalies can be further analyzed and investigated to determine if they are truly anomalous
 or if they represent a new, previously unseen pattern in the data.
```
